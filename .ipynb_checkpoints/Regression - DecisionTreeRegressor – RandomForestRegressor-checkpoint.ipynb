### 1.3. Répartition des données

```python
X = df.drop('median_house_value', axis=1)
Y= df['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

x_test.head()
```

### 1.4. Découverte et visualisation des données

```python
plt.scatter(x_train['longitude'], x_train['latitude'], alpha=0.1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
image = plt.imread('california.png')
fig, ax = plt.subplots(figsize = (10,7))

ax.imshow(image,zorder=0,extent=[df['longitude'].min(), df['longitude'].max(), df['latitude'].min(), df['latitude'].max()])

x_train.plot(ax=ax, kind='scatter', x='longitude', y='latitude', alpha=0.8, s=x_train['population']/100, label='population', 
             c=y_train, cmap=plt.get_cmap('RdPu'), colorbar=True, sharex=False, figsize=(12,8), zorder=1)
```
